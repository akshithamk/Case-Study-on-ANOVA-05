# #● Check whether all the zones generate the same amount of sales.

Null Hypothesis:Same amount was generated by all the zones

Alternate Hypothesis:Not the same amount was  generated by all the zones

Independent variables are also termed as factor. For this experimental design, there is only 1 factor or independent variable to evaluate[Amount of Sales], and therefore, one-way ANOVA method is suitable for analysis.

```python
from scipy.stats import f_oneway
```

```python
import scipy.stats as stats
```

```python
f_value, p_value = stats.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'], data['Zone - D'])
```

```python
print(float(p_value))
```

```python
print(float(f_value))
```

insights:consider alpha=0.05,since p _value is much less than alpha,we reject null hypothesis.and accept alternate hypothesis
